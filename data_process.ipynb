```python
import pandas as pd
meta = pd.read_csv("metadata.csv")
chargedata = meta.loc[meta["type"]=="charge"]
dischargedata = meta.loc[meta["type"]=="discharge"]
impedancedata = meta.loc[meta["type"]=="impedance"]
```

```python
import shutil
import os
import glob

battery_id = chargedata.groupby("battery_id")
for bg in ["B0005"]:
    battery = battery_id.get_group(bg)
    cycle = 0
    file = meta[meta["battery_id"]==bg]
    # print(bg)
    for csvname in battery["filename"]:
        it = meta[meta["filename"]==csvname].index
        if it-1>min(file.index):
            nx = file.loc[it-1,["type"]]
            if (nx["type"] == "charge").bool():
                print(file.loc[it,["filename"]])
                continue
        if not os.path.isdir(f"processed_data/charge/{bg}"):
            os.makedirs(f"processed_data/charge/{bg}")
        print(csvname)
        df_tmp = pd.read_csv(f"data/{csvname}")
        cycle+=1
        df_tmp.to_csv(f"processed_data/charge/{bg}/{bg}_charge_cycle_{cycle}.csv")
        # shutil.copy(f"data/{csvname}",f"processed_data/charge/{bg}/{bg}_charge_cycle_{cycle}.csv")
        # shutil.
        
```

```python
battery_id = dischargedata.groupby("battery_id")
for bg in battery_id.groups:
    battery = battery_id.get_group(bg)
    file = meta[meta["battery_id"]==bg]
    cycle = 1
    for csvname in battery["filename"]:
        it = meta[meta["filename"]==csvname].index
        if it-1>min(file.index):
            nx = file.loc[it-1,["type"]]
            if (nx["type"] == "discharge").bool():
                print(file.loc[it,["filename"]])
                continue
        if not os.path.isdir(f"processed_data/discharge/{bg}"):
            os.makedirs(f"processed_data/discharge/{bg}")
        shutil.copyfile(f"data/{csvname}",f"processed_data/discharge/{bg}/{bg}_discharge_cycle_{cycle}.csv")
        cycle+=1
```

```python
battery_id = impedancedata.groupby("battery_id")
for bg in battery_id.groups:
    battery = battery_id.get_group(bg)
    cycle = 1
    for csvname in battery["filename"]:
        if not os.path.isdir(f"processed_data/impedance/{bg}"):
            os.makedirs(f"processed_data/impedance/{bg}")
        shutil.copyfile(f"data/{csvname}",f"processed_data/impedance/{bg}/{bg}_impedance_cycle_{cycle}.csv")
        cycle+=1
```
